# Matplotlib and Seaborn Visualizations
Using the Iris dataset from seaborn


```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
# Load Iris dataset from seaborn
df = sns.load_dataset('iris')
df.head()
```

## 1. Line Plot (Matplotlib)

```python
plt.figure(figsize=(8, 5))
plt.plot(df['sepal_length'][:30])
plt.xlabel('Sample')
plt.ylabel('Sepal Length (cm)')
plt.title('Sepal Length Across Samples')
plt.show()
```

## 2. Scatter Plot (Matplotlib)

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['sepal_length'], df['sepal_width'])
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.title('Sepal Length vs Width')
plt.show()
```

## 3. Histogram (Seaborn)

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['petal_length'], bins=20, kde=True)
plt.xlabel('Petal Length (cm)')
plt.title('Distribution of Petal Length')
plt.show()
```

## 4. Box Plot (Seaborn)

```python
plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='species', y='petal_width')
plt.xlabel('Species')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Width by Species')
plt.show()
```
